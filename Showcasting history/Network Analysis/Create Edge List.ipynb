```python
import pandas as pd
```

```python
df = pd.read_csv("Analysis_version.csv")
df.head()
```

# Network - People Contributed on the Same Work

```python
df_people = df['ID'], df['Personal Name']
df_people
```

```python
people_dict = {}
i = 0
for ID in df_people[0]:
    string = str(df_people[1][i]).replace('\xa0', ' ').strip()
    if ID not in people_dict:
        people_dict[ID] = [str(string)]
    else:
        people_dict[ID].append(str(string))
    i += 1
people_dict
```

```python
raw_list = [people_dict[ID] for ID in people_dict]
name_list = []
for group in raw_list:
    without_nan = []
    for name in group:
        if not name == 'nan':
            without_nan.append(name)
    name_list.append(without_nan)
group_list = [group for group in name_list if len(group) > 1]
group_list
```

```python
people_edges = []
duplicated_list = []
for item in group_list:
    pairs = [[a, b] for idx, a in enumerate(item) for b in item[idx + 1:]]
    for pair in pairs:
        if not pair[0] == pair[1]:
            duplicated_list.append(pair)
for pair in duplicated_list:
    if pair not in people_edges:
        people_edges.append(pair)
people_edges
```

```python
import csv
with open('Edge_List_Contributers.csv', 'w', encoding = 'utf-8', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(['Source', 'Target'])
    for pair in people_edges:
        writer.writerow(pair)
```

# Network - Publishers Published the Same Authors

```python
df_publishers = df['Personal Name'], df['Publisher Name'], (df['Relationship'] == 'aut')
df_publishers
```

```python
publisher_dict = {}
i = 0
for ID in df_publishers[0]:
    if df_publishers[2][i] == True:
        ID_clean = str(ID).replace('\xa0', ' ').strip()
        string = str(df_publishers[1][i]).replace('\xa0', ' ').strip()
        if ID_clean not in publisher_dict:
            publisher_dict[ID_clean] = [string]
        else:
            publisher_dict[ID_clean].append(string)
    i += 1
publisher_dict
```

```python
raw_list = [publisher_dict[ID] for ID in publisher_dict]
name_list = []
for group in raw_list:
    without_nan = []
    for name in group:
        if not name == 'nan':
            without_nan.append(name)
    name_list.append(without_nan)
group_list = [group for group in name_list if len(group) > 1]
group_list
```

```python
publishers_edges = []
duplicated_list = []
for item in group_list:
    pairs = [[a, b] for idx, a in enumerate(item) for b in item[idx + 1:]]
    for pair in pairs:
        if not pair[0] == pair[1]:
            duplicated_list.append(pair)
for pair in duplicated_list:
    if pair not in publishers_edges:
        publishers_edges.append(pair)
publishers_edges
```

```python
import csv
with open('Edge_List_Publishers.csv', 'w', encoding = 'utf-8', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(['Source', 'Target'])
    for pair in publishers_edges:
        writer.writerow(pair)
```
